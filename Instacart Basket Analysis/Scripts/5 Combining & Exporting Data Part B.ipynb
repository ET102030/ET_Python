## Combining orders_products_combined data set with df_prods_checked

```python
## 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

##Importing products data set from the prepared data folder

df_prods_checked = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', '4.5 prods_checked.csv'), index_col=False)
df_prods_checked.head()

```

```python
df_prods_checked['prices'].max()
```

```python
## These two data frames share the product_id columns as a key so that will be used throughout the merge process

df_orders_prod_merged = df_orders_products_combined.merge(df_prods_checked, on = 'product_id', indicator = 'match')

df_orders_prod_merged.head()
```

```python
## 6. Confirm the results of the merge using the merge flag.

df_orders_prod_merged['_merge'].value_counts()

#There are 0 values of left/right's only inferring everything has been matched accordingly. If this was a different join like an "outer" there may be a larger amonut of values that did not match fully. The inner join nature allows proper matching throughout the merge
```

```python
# drop the flag column '_merge' from the previous merge, along with unnamed columns for data to look more "clean"

orders_prod_merged_dropped = df_orders_prod_merged.drop(['match', '_merge', 'Unnamed: 0.1', 'Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1 )

orders_prod_merged_dropped.head()
## New DF "orders_prod_merged_dropped" made to prevent over writting in case of error
```

```python
## 7. Export the newly created dataframe as ords_prods_merge in a suitable format (taking into consideration the size).

## Will use pickle format due to file size

orders_prod_merged_dropped.to_pickle(os.path.join(path, '02 Data','Prepared Data', '4.6 orders_prod_merged.pkl'))

## Will also use csv in case needed in the future
orders_prod_merged_dropped.to_csv(os.path.join(path, '02 Data','Prepared Data', '4.6 orders_prod_merged.csv'))
```
