# 5. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

```python
#will need to randomly sample the data, and then generate a scatter plot chart
np.random.seed(4)
dev=np.random.rand(len(customer_ords_prods))<=0.7
dev
#boolean array has been set up for a 70:30 randomized sample
```

```python
#Storing 70% of the sample into big
big_market = customer_ords_prods[dev]

#Storing the remaining 30% of hte sample into small
small_market = customer_ords_prods[~dev]
```

```python
#comparing it to big+small
len(big_market)+len(small_market)

# we can see they are equal, so it was done correctly
```

```python
#Checking to see if the split was correct using the len() statement

len(customer_ords_prods)
```

```python
#Reducing the sample to only those columns necessary for the chart

df_small_hour_price = small_market[['order_time_hour_of_day','prices']]
```

```python
#Using df_small_hour_price
line_hour_price = sns.lineplot(data = df_small_hour_price, x = 'order_time_hour_of_day',y = 'prices')
```

```python
line_hour_price.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_hod_price.png'))
```
