```python
#Checking for mixed data

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)

#Reports only first_name is providing mixed data, with it's Nan 
```

```python
#Subsetting that First Name column based on the NaN values

df_first_name = customers[customers['first_name'].isnull()==True]
df_first_name.head()
```
